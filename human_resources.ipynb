# Import the Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
employee_df = pd.read_csv("Human_Resources.csv")
```

# Data Exploration and Transformation

```python
employee_df.info()
```

This DataFrame represents employee data from a company.

- The "Attrition" column indicates whether an employee has left the company or not.
- There are 1470 entries and 35 columns in the DataFrame.
Data types include integers and objects (likely representing categorical variables).

```python
employee_df.describe()
```

```python
# Convert categorical variables to binary representations.

employee_df["Attrition"] = employee_df["Attrition"].apply(
    lambda x: 1 if x == "Yes" else 0
)
employee_df["OverTime"] = employee_df["OverTime"].apply(
    lambda x: 1 if x == "Yes" else 0
)
employee_df["Over18"] = employee_df["Over18"].apply(lambda x: 1 if x == "Y" else 0)
```

```python
# Check for the missing values using a heatmap.

sns.heatmap(employee_df.isnull(), yticklabels=False, cbar=False, cmap="Blues")
```

As we can see, there are no missing values.

```python
# Creating histogram plots of the attributes to visualise the data.

employee_df.hist(bins=30, figsize=(20, 20), color="r")
```

```python
# Retrieve the list of column names from the DataFrame and print the total number of columns.
columns = employee_df.columns.to_list()
print(len(columns))

# Define a list of columns to be deleted.
delete_cols = [
    "EmployeeCount",
    "StandardHours",
    "Over18",
    "EmployeeNumber",
    "Gender",
    "Department",
    "BusinessTravel",
    "MaritalStatus",
    "JobRole",
    "EducationField",
]

# Filter out the columns specified in 'delete_cols' from the original list of columns
columns = [col for col in columns if col not in delete_cols]

# Print the updated number of columns after removing the specified ones
print(len(columns))
```

- The DataFrame, initially consisting 35 columns. 
- After removing specified columns we got reduced set of 25 columns.

```python
# How many employees stayed and how many left the organisation.

left_df = employee_df[employee_df["Attrition"] == 1]
stayed_df = employee_df[employee_df["Attrition"] == 0]
```

```python
total = len(employee_df)

left = len(left_df)
left_percentage = round(left / total * 100)

stayed = len(stayed_df)
stayed_percentage = round(stayed / total * 100)

print(f"left: {left} ({left_percentage} %)")
print(f"stayed: {stayed} ({stayed_percentage} %)")
```

```python
# Let's compare the average number of employees who stayed and left the organisation.

left_df.describe()
```

```python
stayed_df.describe()
```

The provided observations compare various factors between employees who chose to stay with the company and those who decided to leave.

1) Age: The average age of the employees who remained with the company is higher compared to those who left. This could indicate that older employees are more likely to stay with the company, possibly due to factors like job security, benefits, or attachment to the workplace.

2) DailyRate: The daily rate of the employees who stayed is higher than those who left. This suggests that higher compensation might be a factor in retaining employees.

3) Distance From Home: Employees who decided to stay live closer to the workplace compared to those who left. A shorter commute might contribute to higher employee retention, as it can reduce stress and increase work-life balance.

4) Environment Satisfaction & Job Satisfaction: Employees who stayed are generally more satisfied with both their work environment and their job. This indicates that job satisfaction and a positive work environment are important factors in retaining employees.

5) Stock Option Level: Employees who chose to stay tend to have higher stock option levels. This suggests that offering stock options as part of compensation might incentivize employees to stay with the company, as they have a vested interest in its success.

```python
# Creating a Correlation Heatmap.

correlations = employee_df[columns].corr()
f, ax = plt.subplots(figsize=(20, 15))
sns.heatmap(correlations, annot=True)
```

Key highlights:

1) Strong correlation between total working years and job level: This suggests that employees who have worked for a longer duration tend to have higher job levels within the organisation which may indicate that promotions or advancements within the company are often based on tenure or accumulated experience.

2) Positive correlation between total working years and monthly income: Employees with more years of experience tend to have higher monthly incomes. This correlation is expected, as experience often translates to expertise and higher-value contributions to the organization, which are typically rewarded with increased compensation.

3) Positive correlation between job level and monthly income: Similarly, there's a positive correlation between job level and monthly income. This indicates that higher-ranking positions within the company are associated with higher salaries or incomes. This correlation aligns with common organisational structures where higher positions command higher compensation.

4) Strong correlation between age and monthly income: The observation suggests that age is strongly correlated with monthly income. This correlation implies that as individuals age, their income tends to increase. 

```python
# Creating a countplot for the Employees who Stayed and who Left based on their Age.

plt.figure(figsize=[20, 12])
sns.countplot(x="Age", hue="Attrition", data=employee_df)
```

These are the highlights of the observations:

Early Career (Early 20s): Employees in their early 20s are often starting their professional journey. They are generally loyal to their organisation as they seek to gain experience, build skills, and establish themselves in their chosen field. They may prioritise growth opportunities over immediate compensation.

Approaching 30s and Mid-Thirties: As employees approach their 30s, particularly in their mid-thirties, they may start considering leaving their current organisation. This could be driven by various factors such as seeking higher wages to support themselves and their families, pursuing better career opportunities elsewhere, or feeling stagnant in their current roles.

Trend Down in Forties: The trend tends to stabilize or even decrease as employees enter their forties. By this age, individuals may have established themselves in their careers, achieved higher positions or salaries, and settled into their personal lives. They may become less inclined to switch jobs frequently, especially if they feel satisfied with their current situation and are closer to retirement age.

```python
# Let's create countplot to show the distribution of the "Attrition" for attributes job role, marital status, job involvement, and job level.

plt.figure(figsize=[20, 20])

plt.subplot(411)
sns.countplot(x="JobRole", hue="Attrition", data=employee_df)

plt.subplot(412)
sns.countplot(x="MaritalStatus", hue="Attrition", data=employee_df)

plt.subplot(413)
sns.countplot(x="JobInvolvement", hue="Attrition", data=employee_df)

plt.subplot(414)
sns.countplot(x="JobLevel", hue="Attrition", data=employee_df)
```

Key highlights:

1) Marital Status: Married individuals tend to stay with the organisation more than single employees and those who are divorced. This could be due to married individuals having more stability and potentially seeking long-term employment for financial security and stability for their families.

2) Job Roles: Sales representatives and research scientists are more likely to leave compared to other job roles. This could be because sales roles often involve high-pressure targets and performance expectations, while research scientists may leave in pursuit of more challenging research opportunities or better compensation elsewhere.

3) Employee Engagement: Less involved employees, who may be less engaged or committed to their work, are more likely to leave the company. This suggests that employee engagement and involvement in work activities play a significant role in retention.

4) Married Employees: The percentage of married individuals is more likely to stay with the organisation. This aligns with the stability and support that marriage can provide, leading to a greater commitment to their current employment.

5) Experience and Promotions: Less experienced employees, typically at lower job levels, are more likely to leave the company. However, as employees are promoted and gain more experience, the likelihood of them leaving gradually decreases. This indicates that career progression and opportunities for advancement within the organisation are crucial factors in retaining employees.

```python
# Kernel Density Estimate probability density

plt.figure(figsize=(12, 7))

sns.kdeplot(
    left_df["DistanceFromHome"], label="Employee who left", shade=True, color="r"
)
sns.kdeplot(
    stayed_df["DistanceFromHome"], label="Employee who stayed", shade=True, color="b"
)

plt.legend()
plt.xlabel("Distance From Home")
```

Employees who stays closer to the workplace are less like to leave.

```python
# Kernel Density Estimate probability density

plt.figure(figsize=(12, 7))

sns.kdeplot(
    left_df["YearsWithCurrManager"], label="Employee who left", shade=True, color="r"
)
sns.kdeplot(
    stayed_df["YearsWithCurrManager"],
    label="Employee who stayed",
    shade=True,
    color="b",
)

plt.legend()
plt.xlabel("Years with Current Manager")
```

Key highlights:

1) Initial Peak (Around 1 Year): Initially, there is a peak in the number of employees leaving the organisation after being with the same manager for about a year. This suggests that within the first year of working under a particular manager, there's a relatively high turnover rate. 

However, after this initial period, the difference in turnover rates between employees who stayed and those who left diminishes, although the number of employees who left remains higher.

2) Secondary Peak (Around 6-7 Years): There is a secondary peak in the data after employees have been with the same manager for around 6-7 years. However, this peak represents a different trend. Employees who have stayed with the same manager for over 5 years are less likely to leave the organisation compared to earlier in their tenure. This suggests that after a certain threshold of time spent with a manager (in this case, around 7 years), the likelihood of turnover decreases significantly. 

This could be attributed to factors such as trust, and familiarity developed between the employee and the manager over time, leading to increased retention.

```python
# Kernel Density Estimate probability density

plt.figure(figsize=(12, 7))

sns.kdeplot(
    left_df["TotalWorkingYears"], label="Employee who left", shade=True, color="r"
)
sns.kdeplot(
    stayed_df["TotalWorkingYears"], label="Employee who stayed", shade=True, color="b"
)

plt.legend()
plt.xlabel("Total Working Years")
```

Initially, the count of employees who stayed and left does not differ significantly, with those who stayed being slightly higher. However, after approximately 15 years of working at the same organisation, the number of employees who left surpasses those who stayed.

## Create Testing and Training Dataset & Perform Data Cleaning

```python
employee_df.head(3)
```

```python
X_cat = employee_df[
    [
        "BusinessTravel",
        "Department",
        "EducationField",
        "Gender",
        "JobRole",
        "MaritalStatus",
    ]
]
```

```python
X_cat
```

```python
from sklearn.preprocessing import OneHotEncoder

onehotencoder = OneHotEncoder()
X_cat = onehotencoder.fit_transform(X_cat).toarray()
```

```python
X_cat
```

```python
X_cat.shape
```

```python
X_cat = pd.DataFrame(X_cat)
```

```python
X_cat
```

```python
employee_df.info()
```

```python
X_numerical = employee_df[
    [
        "Age",
        "DailyRate",
        "DistanceFromHome",
        "Education",
        "EmployeeCount",
        "EmployeeNumber",
        "EnvironmentSatisfaction",
        "HourlyRate",
        "JobInvolvement",
        "JobLevel",
        "JobSatisfaction",
        "MonthlyIncome",
        "MonthlyRate",
        "NumCompaniesWorked",
        "PercentSalaryHike",
        "PerformanceRating",
        "RelationshipSatisfaction",
        "StandardHours",
        "StockOptionLevel",
        "TotalWorkingYears",
        "TrainingTimesLastYear",
        "WorkLifeBalance",
        "YearsAtCompany",
        "YearsInCurrentRole",
        "YearsSinceLastPromotion",
        "YearsWithCurrManager",
        "OverTime",
    ]
]
X_numerical.info()
```

```python
X_all = pd.concat([X_cat, X_numerical], axis=1)

X_all
```

```python
X_all.columns
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X = scaler.fit_transform(X_all.values)

X
```

```python
y = employee_df["Attrition"]

y
```

## Logistic Regression Classifier

## 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print("Accuracy {} %".format(100 * accuracy_score(y_pred, y_test)))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(
    model, X, y, display_labels=["Stayed", "Left"], cmap="Greens"
)
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Artificial Neural Network

```python
import tensorflow as tf

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=500, activation="relu", input_shape=(53,)))
model.add(tf.keras.layers.Dense(units=500, activation="relu"))
model.add(tf.keras.layers.Dense(units=500, activation="relu"))
model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(optimizer="Adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
epochs_hist = model.fit(X_train, y_train, epochs=100, batch_size=50)
```

```python
y_pred = model.predict(X_test)
y_pred = y_pred > 0.5

y_pred
```

```python
plt.plot(epochs_hist.history["loss"])
plt.title("Model Loss Progresss During Training")
plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.legend(["Training Loss"])
```

```python
plt.plot(epochs_hist.history["accuracy"])
plt.title("Model Accuracy Progresss During Training")
plt.xlabel("Epoch")
plt.ylabel("Training Accuracy")
plt.legend(["Training Accuracy"])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

accuracy = 100 * accuracy_score(y_pred, y_test)
print(f"Accuracy {round(accuracy,2)} %")

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Stayed", "Left"])
disp.plot(cmap="Blues")
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```
